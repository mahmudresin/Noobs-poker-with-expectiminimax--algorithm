```python
import pygame
import random
import os

# Initialize Pygame
pygame.init()

# Constants
SCREEN_WIDTH, SCREEN_HEIGHT = 800, 600
CARD_WIDTH, CARD_HEIGHT = 72, 96
WHITE = (255, 255, 255)
GREEN = (0, 128, 0)

# Path to the card images
card_images_path = '/content/drive/MyDrive/PNG-cards-1.3/PNG-cards-1.3'

# Load the card images
deck_images = []
for i in range(52):
    image_path = os.path.join(card_images_path, f'{i}.png')
    try:
        deck_images.append(pygame.image.load(image_path))
    except FileNotFoundError:
        print(f"Error: The image file '{image_path}' was not found.")

# Check if all images are loaded
print(f"Total images loaded: {len(deck_images)}")

if len(deck_images) != 52:
    print("Warning: Not all card images were loaded. Check your image files and path.")
else:
    print("All card images loaded successfully.")

# Setup screen
screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))
pygame.display.set_caption('Poker Game')

# Define the Card class
class Card:
    def __init__(self, suit, rank):
        self.suit = suit
        self.rank = rank
        try:
            self.image = deck_images[rank + suit * 13]  # Correct index for card images
        except IndexError:
            print(f"Error: Invalid index {rank + suit * 13} for suit {suit}, rank {rank}.")
            self.image = None

    def draw(self, x, y):
        if self.image:
            screen.blit(self.image, (x, y))

# Poker logic
class PokerGame:
    def __init__(self):
        self.deck = [Card(suit, rank) for suit in range(4) for rank in range(13)]
        self.community_cards = []
        self.player_hand = []
        self.ai_hand = []
        self.player_money = 1000
        self.ai_money = 1000

    def deal(self):
        random.shuffle(self.deck)
        self.player_hand = [self.deck.pop() for _ in range(2)]
        self.ai_hand = [self.deck.pop() for _ in range(2)]
        self.community_cards = [self.deck.pop() for _ in range(5)]

    def draw_table(self):
        screen.fill(GREEN)
        # Draw community cards
        for i, card in enumerate(self.community_cards):
            card.draw(200 + i * (CARD_WIDTH + 10), 200)
        # Draw player cards
        for i, card in enumerate(self.player_hand):
            card.draw(100 + i * (CARD_WIDTH + 10), 400)

# Initialize the game
poker_game = PokerGame()
poker_game.deal()

# Game loop
running = True
while running:
    screen.fill(WHITE)

    poker_game.draw_table()

    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    pygame.display.flip()

pygame.quit()
```
